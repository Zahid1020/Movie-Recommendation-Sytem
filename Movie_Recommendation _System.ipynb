```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
movies.head(3)
```

```python
print(movies.shape)
print(credits.shape)
```

```python
movies = movies.merge(credits, on = 'title')
movies.shape
```

```python
movies.info()
```

```python
df = movies[['genres','id', 'keywords', 'overview', 'title', 'cast', 'crew']]
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.head(3)
```

```python
import ast
```

```python
def conv(object):
    L =[]
    for i in ast.literal_eval(object):
        L.append(i['name'])
    return L
```

```python
df['genres'] = df['genres'].apply(conv)
```

```python
df['keywords'] = df['keywords'].apply(conv)
```

```python
def conv3(obj):
    L= [] 
    c= 0
    for i in ast.literal_eval(obj):
        if c != 3:
            L.append(i['name'])
            c+=1
        else:
            break
    return L
```

```python
df['cast'] = df['cast'].apply(conv3)
```

```python
def director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
    
```

```python
df['crew'] = df['crew'].apply(director)
```

```python
df.head(3)
```

```python
df['overview'] = df['overview'].apply(lambda x: x.split())
```

```python
df['overview']
```

```python
df['genres'] = df['genres'].apply(lambda x: [i.replace(' ', '') for i in x])
df['cast'] = df['cast'].apply(lambda x: [i.replace(' ', '') for i in x])
df['crew'] = df['crew'].apply(lambda x: [i.replace(' ', '') for i in x])
```

```python
df.head(3)
```

```python
df['tags'] = df['genres'] + df['keywords'] + df['overview'] + df['cast'] + df['crew'] 
df = df.drop(['genres', 'keywords', 'overview', 'cast', 'crew'], axis=1)
df.head(3)
```

```python
df['tags'] = df['tags'].apply(lambda x: " ".join(x))
df['tags'] = df['tags'].apply(lambda x: x.lower())
df.head(2)
```

```python
df['tags'][0]
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(obj):
    y = []
    for i in obj.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
len(cv.get_feature_names())
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity.shape
```

```python
similarity[1]
```

```python
def recommend(movie):
    movie_index = df[df["title"] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
